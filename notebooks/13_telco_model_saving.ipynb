# **Telecom Customer Churn - Model Saving**

```python
# =========================================
# 1. Import Required Libraries
# =========================================

import pandas as pd
import numpy as np

import joblib
```

```python
# =========================================
# 2. Import Final Selected Model
# =========================================

from sklearn.naive_bayes import GaussianNB
```

```python
# =========================================
# 3. Load Training Dataset
# =========================================

X_train = pd.read_csv(
'https://raw.githubusercontent.com/Saiful-2/datasets/main/X_train_scaled.csv'
)

y_train = pd.read_csv(
'https://raw.githubusercontent.com/Saiful-2/datasets/main/y_train.csv'
)
```

```python
# =========================================
# 4. Convert Target Variable into Series
# =========================================

y_train = y_train.squeeze()
```

```python
# =========================================
# 5. Initialize Final Model
# =========================================

model = GaussianNB()
```

```python
# =========================================
# 6. Train Final Model
# =========================================

model.fit(

X_train,

y_train

)
```

```python
# =========================================
# 7. Save Trained Model
# =========================================

joblib.dump(

model,

'churn_model.pkl'

)
```

```python
# =========================================
# 8. Verify Model Saved
# =========================================

loaded_model = joblib.load(

'churn_model.pkl'

)

loaded_model
```

```python
# =========================================
# 9. Save Feature Names
# =========================================

feature_names = X_train.columns

joblib.dump(

feature_names,

'feature_names.pkl'

)
```

```python
# =========================================
# 10. Verify Feature Names
# =========================================

loaded_features = joblib.load(

'feature_names.pkl'

)

loaded_features
```

```python
# =========================================
# 11. Save Scaler
# =========================================

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(

X_train

)

joblib.dump(

scaler,

'scaler.pkl'

)
```

```python
# =========================================
# 12. Verify Scaler
# =========================================

loaded_scaler = joblib.load(

'scaler.pkl'

)

loaded_scaler
```

```python
# =========================================
# 13. Save Model Metadata
# =========================================

metadata = {

'Model':'GaussianNB',

'Target':'Churn',

'Version':'1.0'

}

joblib.dump(

metadata,

'metadata.pkl'

)
```

```python
# =========================================
# 14. Verify Metadata
# =========================================

loaded_metadata = joblib.load(

'metadata.pkl'

)

loaded_metadata
```

```python
# =========================================
# 15. Test Saved Model
# =========================================

sample = X_train.iloc[[0]]

prediction = loaded_model.predict(

sample

)

prediction
```

```python
# =========================================
# 16. Prediction Probability
# =========================================

loaded_model.predict_proba(

sample

)
```

```python
# =========================================
# 17. Save Files Summary
# =========================================

print(

"Saved Files"

)

print(

"1. churn_model.pkl"

)

print(

"2. scaler.pkl"

)

print(

"3. feature_names.pkl"

)

print(

"4. metadata.pkl"

)
```

```python
# =========================================
# 18. Final Conclusion
# =========================================

print(

"Gaussian Naive Bayes model saved successfully."

)
```
