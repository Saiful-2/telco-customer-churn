# Telecom Customer Churn - Data Cleaning

```python
# =========================================
# 1. Import Required Libraries
# =========================================

import pandas as pd
import numpy as np
```

```python
# =========================================
# 2. Load Dataset
# =========================================

url = "https://raw.githubusercontent.com/Saiful-2/datasets/main/telco-customer-churn.csv"

df = pd.read_csv(url)
```

```python
# =========================================
# 4. Display Last 5 Rows
# =========================================

df.tail()
```

```python
# =========================================
# 5. Check Shape of Dataset
# =========================================

df.shape
```

```python
# =========================================
# 6. Display Column Names
# =========================================

df.columns
```

```python
# =========================================
# 7. Dataset Information
# =========================================

df.info()
```

```python
# =========================================
# 8. Check Data Types
# =========================================

df.dtypes
```

```python
# =========================================
# 9. Statistical Summary
# =========================================

df.describe()
```

```python
# =========================================
# 10. Statistical Summary Including Categorical Columns
# =========================================

df.describe(include='all')
```

```python
# =========================================
# 11. Check Missing Values
# =========================================

df.isnull().sum()
```

```python
# =========================================
# 12. Check Blank Spaces in Dataset
# =========================================

(df == ' ').sum()
```

```python
# =========================================
# 13. Check Unique Values of TotalCharges
# =========================================

df['TotalCharges'].unique()
```

```python
# =========================================
# 14. Convert TotalCharges to Numeric
# =========================================

df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
# =========================================
# 15. Check Missing Values Again
# =========================================

df.isnull().sum()
```

```python
# =========================================
# 16. Display Rows with Missing Values
# =========================================

df[df['TotalCharges'].isnull()]
```

```python
# =========================================
# 17. Remove Missing Values
# =========================================

df = df.dropna()
```

```python
# =========================================
# 18. Verify Missing Value Removal
# =========================================

df.isnull().sum()
```

```python
# =========================================
# 19. Check Duplicate Rows
# =========================================

df.duplicated().sum()
```

```python
# =========================================
# 20. Remove Duplicate Rows
# =========================================

df = df.drop_duplicates()
```

```python
# =========================================
# 21. Verify Duplicate Removal
# =========================================

df.duplicated().sum()
```

```python
# =========================================
# 22. Check Unique Values of All Columns
# =========================================

for col in df.columns:
    print("Column Name :", col)
    print(df[col].unique())
    print("-" * 50)
```

```python
# =========================================
# 23. Check Unique Values of Target Column
# =========================================

df['Churn'].unique()
```

```python
# =========================================
# 24. Convert Churn Column into Numeric
# =========================================

df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})
```

```python
# =========================================
# 25. Verify Churn Conversion
# =========================================

df['Churn'].head()
```

```python
# =========================================
# 26. Check Data Types Again
# =========================================

df.dtypes
```

```python
# =========================================
# 27. Check Number of Unique Values
# =========================================

df.nunique()
```

```python
# =========================================
# 28. Remove Unnecessary Column
# =========================================

df = df.drop('customerID', axis=1)
```

```python
# =========================================
# 29. Final Dataset Shape
# =========================================

df.shape
```

```python
# =========================================
# 30. Final Dataset Information
# =========================================

df.info()
```

```python
# =========================================
# 31. Display Cleaned Dataset
# =========================================

df.head()
```

```python
# =========================================
# 32. Save Cleaned Dataset
# =========================================

df.to_csv("cleaned_data.csv", index=False)
```
