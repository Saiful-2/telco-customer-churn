# Telecom Customer Churn - Exploratory Data Analysis

```python
# =========================================
# 1. Import Required Libraries
# =========================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# =========================================
# 2. Load Cleaned Dataset
# =========================================

url = "https://raw.githubusercontent.com/Saiful-2/datasets/main/cleaned-telco-customer-churn.csv"

df = pd.read_csv(url)
```

```python
# =========================================
# 3. Display First 5 Rows
# =========================================

df.head()
```

```python
# =========================================
# 4. Check Dataset Shape
# =========================================

df.shape
```

```python
# =========================================
# 5. Dataset Information
# =========================================

df.info()
```

```python
# =========================================
# 6. Statistical Summary
# =========================================

df.describe()
```

```python
# =========================================
# 7. Statistical Summary Including Categorical Columns
# =========================================

df.describe(include='all')
```

```python
# =========================================
# 8. Check Churn Distribution
# =========================================

df['Churn'].value_counts()
```

```python
# =========================================
# 9. Churn Countplot
# =========================================

sns.countplot(x='Churn', data=df)

plt.title("Customer Churn Distribution")

plt.show()
```

```python
# =========================================
# 10. Churn Percentage Distribution
# =========================================

df['Churn'].value_counts(normalize=True) * 100
```

```python
# =========================================
# 11. Churn Pie Chart
# =========================================

df['Churn'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    figsize=(6,6)
)

plt.title("Customer Churn Percentage")

plt.ylabel("")

plt.show()
```

```python
# =========================================
# 12. Gender Distribution
# =========================================

sns.countplot(x='gender', data=df)

plt.title("Gender Distribution")

plt.show()
```

```python
# =========================================
# 13. Churn vs Gender
# =========================================

sns.countplot(x='gender', hue='Churn', data=df)

plt.title("Churn vs Gender")

plt.show()
```

```python
# =========================================
# 14. Senior Citizen Distribution
# =========================================

sns.countplot(x='SeniorCitizen', data=df)

plt.title("Senior Citizen Distribution")

plt.show()
```

```python
# =========================================
# 15. Churn vs Senior Citizen
# =========================================

sns.countplot(x='SeniorCitizen', hue='Churn', data=df)

plt.title("Churn vs Senior Citizen")

plt.show()
```

```python
# =========================================
# 16. Partner Distribution
# =========================================

sns.countplot(x='Partner', data=df)

plt.title("Partner Distribution")

plt.show()
```

```python
# =========================================
# 17. Churn vs Partner
# =========================================

sns.countplot(x='Partner', hue='Churn', data=df)

plt.title("Churn vs Partner")

plt.show()
```

```python
# =========================================
# 18. Dependents Distribution
# =========================================

sns.countplot(x='Dependents', data=df)

plt.title("Dependents Distribution")

plt.show()
```

```python
# =========================================
# 19. Churn vs Dependents
# =========================================

sns.countplot(x='Dependents', hue='Churn', data=df)

plt.title("Churn vs Dependents")

plt.show()
```

```python
# =========================================
# 20. Contract Type Distribution
# =========================================

sns.countplot(x='Contract', data=df)

plt.title("Contract Type Distribution")

plt.xticks(rotation=20)

plt.show()
```

```python
# =========================================
# 21. Churn vs Contract Type
# =========================================

sns.countplot(x='Contract', hue='Churn', data=df)

plt.title("Churn vs Contract Type")

plt.xticks(rotation=20)

plt.show()
```

```python
# =========================================
# 22. Internet Service Distribution
# =========================================

sns.countplot(x='InternetService', data=df)

plt.title("Internet Service Distribution")

plt.show()
```

```python
# =========================================
# 23. Churn vs Internet Service
# =========================================

sns.countplot(x='InternetService', hue='Churn', data=df)

plt.title("Churn vs Internet Service")

plt.show()
```

```python
# =========================================
# 24. Payment Method Distribution
# =========================================

sns.countplot(x='PaymentMethod', data=df)

plt.title("Payment Method Distribution")

plt.xticks(rotation=45)

plt.show()
```

```python
# =========================================
# 25. Churn vs Payment Method
# =========================================

sns.countplot(x='PaymentMethod', hue='Churn', data=df)

plt.title("Churn vs Payment Method")

plt.xticks(rotation=45)

plt.show()
```

```python
# =========================================
# 26. Tenure Distribution
# =========================================

sns.histplot(df['tenure'], bins=30, kde=True)

plt.title("Tenure Distribution")

plt.show()
```

```python
# =========================================
# 27. Monthly Charges Distribution
# =========================================

sns.histplot(df['MonthlyCharges'], bins=30, kde=True)

plt.title("Monthly Charges Distribution")

plt.show()
```

```python
# =========================================
# 28. Total Charges Distribution
# =========================================

sns.histplot(df['TotalCharges'], bins=30, kde=True)

plt.title("Total Charges Distribution")

plt.show()
```

```python
# =========================================
# 29. Churn vs Tenure
# =========================================

sns.boxplot(x='Churn', y='tenure', data=df)

plt.title("Churn vs Tenure")

plt.show()
```

```python
# =========================================
# 30. Churn vs Monthly Charges
# =========================================

sns.boxplot(x='Churn', y='MonthlyCharges', data=df)

plt.title("Churn vs Monthly Charges")

plt.show()
```

```python
# =========================================
# 31. Churn vs Total Charges
# =========================================

sns.boxplot(x='Churn', y='TotalCharges', data=df)

plt.title("Churn vs Total Charges")

plt.show()
```

```python
# =========================================
# 32. Correlation Matrix
# =========================================

numerical_df = df.select_dtypes(include=['int64', 'float64'])

correlation = numerical_df.corr()

correlation
```

```python
# =========================================
# 33. Correlation Heatmap
# =========================================

plt.figure(figsize=(8,6))

sns.heatmap(correlation, annot=True, cmap='coolwarm')

plt.title("Correlation Heatmap")

plt.show()
```

```python
# =========================================
# 34. Pairplot of Numerical Features
# =========================================

sns.pairplot(numerical_df)

plt.show()
```

```python
# =========================================
# 35. Churn Rate by Contract Type
# =========================================

pd.crosstab(df['Contract'], df['Churn'], normalize='index') * 100
```

```python
# =========================================
# 36. Churn Rate by Internet Service
# =========================================

pd.crosstab(df['InternetService'], df['Churn'], normalize='index') * 100
```

```python
# =========================================
# 37. Churn Rate by Payment Method
# =========================================

pd.crosstab(df['PaymentMethod'], df['Churn'], normalize='index') * 100
```
