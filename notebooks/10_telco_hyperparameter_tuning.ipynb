# **Telecom Customer Churn - Hyperparameter Tuning**

```python
# =========================================
# 1. Import Required Libraries
# =========================================

import pandas as pd
import numpy as np
```

```python
#

!pip install catboost
```

```python
# =========================================
# 2. Import Models
# =========================================

from sklearn.ensemble import RandomForestClassifier

from xgboost import XGBClassifier

from lightgbm import LGBMClassifier

from catboost import CatBoostClassifier

from sklearn.svm import SVC
```

```python
# =========================================
# 3. Import Hyperparameter Tuning Libraries
# =========================================

from sklearn.model_selection import GridSearchCV

from sklearn.model_selection import RandomizedSearchCV
```

```python
# =========================================
# 4. Import Evaluation Metrics
# =========================================

from sklearn.metrics import accuracy_score

from sklearn.metrics import precision_score

from sklearn.metrics import recall_score

from sklearn.metrics import f1_score
```

```python
# =========================================
# 5. Load Training Dataset
# =========================================

X_train = pd.read_csv(
'https://raw.githubusercontent.com/Saiful-2/datasets/main/X_train_scaled.csv'
)

y_train = pd.read_csv(
'https://raw.githubusercontent.com/Saiful-2/datasets/main/y_train.csv'
)
```

```python
# =========================================
# 6. Load Testing Dataset
# =========================================

X_test = pd.read_csv(
'https://raw.githubusercontent.com/Saiful-2/datasets/main/X_test_scaled.csv'
)

y_test = pd.read_csv(
'https://raw.githubusercontent.com/Saiful-2/datasets/main/y_test.csv'
)
```

```python
# =========================================
# 7. Convert Target Variable into Series
# =========================================

y_train = y_train.squeeze()

y_test = y_test.squeeze()
```

```python
# =========================================
# 8. Random Forest Hyperparameter Tuning
# =========================================

rf = RandomForestClassifier(
    random_state=42
)

rf_params = {

'n_estimators':[100,200,300],

'max_depth':[5,10,15],

'min_samples_split':[2,5,10],

'min_samples_leaf':[1,2,4]

}

rf_grid = GridSearchCV(

rf,

rf_params,

cv=5,

scoring='f1',

n_jobs=-1

)

rf_grid.fit(

X_train,

y_train

)
```

```python
# =========================================
# 9. Best Random Forest Parameters
# =========================================

rf_grid.best_params_
```

```python
# =========================================
# 10. Tuned Random Forest Prediction
# =========================================

rf_best = rf_grid.best_estimator_

rf_pred = rf_best.predict(
X_test
)
```

```python
# =========================================
# 11. XGBoost Hyperparameter Tuning
# =========================================

xgb = XGBClassifier(

random_state=42,

eval_metric='logloss'

)

xgb_params = {

'n_estimators':[100,200],

'max_depth':[3,5,7],

'learning_rate':[0.01,0.05,0.1],

'subsample':[0.8,1.0]

}

xgb_random = RandomizedSearchCV(

xgb,

xgb_params,

n_iter=10,

cv=5,

scoring='f1',

random_state=42,

n_jobs=-1

)

xgb_random.fit(

X_train,

y_train

)
```

```python
# =========================================
# 12. Best XGBoost Parameters
# =========================================

xgb_random.best_params_
```

```python
# =========================================
# 13. Tuned XGBoost Prediction
# =========================================

xgb_best = xgb_random.best_estimator_

xgb_pred = xgb_best.predict(
X_test
)
```

```python
# =========================================
# 14. LightGBM Hyperparameter Tuning
# =========================================

lgbm = LGBMClassifier(
random_state=42
)

lgbm_params = {

'n_estimators':[100,200],

'max_depth':[5,10],

'learning_rate':[0.01,0.1],

'num_leaves':[31,50]

}

lgbm_random = RandomizedSearchCV(

lgbm,

lgbm_params,

n_iter=8,

cv=5,

scoring='f1',

random_state=42,

n_jobs=-1

)

lgbm_random.fit(

X_train,

y_train

)
```

```python
# =========================================
# 15. Best LightGBM Parameters
# =========================================

lgbm_random.best_params_
```

```python
# =========================================
# 16. Tuned LightGBM Prediction
# =========================================

lgbm_best = lgbm_random.best_estimator_

lgbm_pred = lgbm_best.predict(
X_test
)
```

```python
# =========================================
# 17. CatBoost Hyperparameter Tuning
# =========================================

cat = CatBoostClassifier(

verbose=0,

random_state=42

)

cat_params = {

'depth':[4,6,8],

'learning_rate':[0.01,0.05,0.1],

'iterations':[100,200]

}

cat_random = RandomizedSearchCV(

cat,

cat_params,

n_iter=8,

cv=5,

scoring='f1',

random_state=42,

n_jobs=-1

)

cat_random.fit(

X_train,

y_train

)
```

```python
# =========================================
# 18. Best CatBoost Parameters
# =========================================

cat_random.best_params_
```

```python
# =========================================
# 19. Tuned CatBoost Prediction
# =========================================

cat_best = cat_random.best_estimator_

cat_pred = cat_best.predict(
X_test
)
```

```python
# =========================================
# 20. SVM Hyperparameter Tuning
# =========================================

svm = SVC(
probability=True
)

svm_params = {

'C':[0.1,1,10],

'kernel':['linear','rbf'],

'gamma':['scale','auto']

}

svm_grid = GridSearchCV(

svm,

svm_params,

cv=5,

scoring='f1',

n_jobs=-1

)

svm_grid.fit(

X_train,

y_train

)
```

```python
# =========================================
# 21. Best SVM Parameters
# =========================================

svm_grid.best_params_
```

```python
# =========================================
# 22. Tuned SVM Prediction
# =========================================

svm_best = svm_grid.best_estimator_

svm_pred = svm_best.predict(
X_test
)
```

```python
# =========================================
# 23. Create Evaluation Function
# =========================================

def evaluate(

name,

y_true,

y_pred

):

    return [

    name,

    accuracy_score(
    y_true,
    y_pred
    ),

    precision_score(
    y_true,
    y_pred
    ),

    recall_score(
    y_true,
    y_pred
    ),

    f1_score(
    y_true,
    y_pred
    )

]
```

```python
# =========================================
# 24. Tuned Model Performance Table
# =========================================

results = []

results.append(
evaluate(
'Tuned Random Forest',
y_test,
rf_pred
)
)

results.append(
evaluate(
'Tuned XGBoost',
y_test,
xgb_pred
)
)

results.append(
evaluate(
'Tuned LightGBM',
y_test,
lgbm_pred
)
)

results.append(
evaluate(
'Tuned CatBoost',
y_test,
cat_pred
)
)

results.append(
evaluate(
'Tuned SVM',
y_test,
svm_pred
)
)

tuned_results = pd.DataFrame(

results,

columns=[

'Model',

'Accuracy',

'Precision',

'Recall',

'F1 Score'

]

)

tuned_results
```

```python
# =========================================
# 25. Rank Models
# =========================================

tuned_results.sort_values(

'F1 Score',

ascending=False

)
```

```python
# =========================================
# 26. Final Conclusion
# =========================================

print(
'Hyperparameter tuning completed successfully.'
)
```
