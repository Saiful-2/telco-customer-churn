# **Telecom Customer Churn - Baseline Models**

```python
# =========================================
# 1. Import Required Libraries
# =========================================

import pandas as pd
import numpy as np
```

```python
# =========================================
# 2. Import Machine Learning Libraries
# =========================================

from sklearn.linear_model import LogisticRegression

from sklearn.tree import DecisionTreeClassifier

from sklearn.neighbors import KNeighborsClassifier

from sklearn.naive_bayes import GaussianNB
```

```python
# =========================================
# 3. Import Evaluation Libraries
# =========================================

from sklearn.metrics import accuracy_score

from sklearn.metrics import precision_score

from sklearn.metrics import recall_score

from sklearn.metrics import f1_score

from sklearn.metrics import confusion_matrix

from sklearn.metrics import classification_report
```

```python
# =========================================
# 4. Load Training Dataset
# =========================================

X_train = pd.read_csv(
    "https://raw.githubusercontent.com/Saiful-2/datasets/main/X_train_scaled.csv"
)

y_train = pd.read_csv(
    "https://raw.githubusercontent.com/Saiful-2/datasets/main/y_train.csv"
)
```

```python
# =========================================
# 5. Load Testing Dataset
# =========================================

X_test = pd.read_csv(
    "https://raw.githubusercontent.com/Saiful-2/datasets/main/X_test_scaled.csv"
)

y_test = pd.read_csv(
    "https://raw.githubusercontent.com/Saiful-2/datasets/main/y_test.csv"
)
```

```python
# =========================================
# 6. Convert Target Variable into Series
# =========================================

y_train = y_train.squeeze()

y_test = y_test.squeeze()
```

```python
# =========================================
# 7. Check Dataset Shapes
# =========================================

print(X_train.shape)

print(X_test.shape)

print(y_train.shape)

print(y_test.shape)
```

```python
# =========================================
# 8. Logistic Regression Model
# =========================================

lr = LogisticRegression(
    random_state=42,
    max_iter=1000
)

lr.fit(X_train, y_train)
```

```python
# =========================================
# 9. Logistic Regression Prediction
# =========================================

lr_pred = lr.predict(X_test)
```

```python
# =========================================
# 10. Logistic Regression Evaluation
# =========================================

print("Accuracy:",
      accuracy_score(y_test, lr_pred))

print("Precision:",
      precision_score(y_test, lr_pred))

print("Recall:",
      recall_score(y_test, lr_pred))

print("F1 Score:",
      f1_score(y_test, lr_pred))
```

```python
# =========================================
# 11. Decision Tree Model
# =========================================

dt = DecisionTreeClassifier(
    random_state=42
)

dt.fit(X_train, y_train)
```

```python
# =========================================
# 12. Decision Tree Prediction
# =========================================

dt_pred = dt.predict(X_test)
```

```python
# =========================================
# 13. Decision Tree Evaluation
# =========================================

print("Accuracy:",
      accuracy_score(y_test, dt_pred))

print("Precision:",
      precision_score(y_test, dt_pred))

print("Recall:",
      recall_score(y_test, dt_pred))

print("F1 Score:",
      f1_score(y_test, dt_pred))
```

```python
# =========================================
# 14. KNN Model
# =========================================

knn = KNeighborsClassifier(
    n_neighbors=5
)

knn.fit(X_train, y_train)
```

```python
# =========================================
# 15. KNN Prediction
# =========================================

knn_pred = knn.predict(X_test)
```

```python
# =========================================
# 16. KNN Evaluation
# =========================================

print("Accuracy:",
      accuracy_score(y_test, knn_pred))

print("Precision:",
      precision_score(y_test, knn_pred))

print("Recall:",
      recall_score(y_test, knn_pred))

print("F1 Score:",
      f1_score(y_test, knn_pred))
```

```python
# =========================================
# 17. Naive Bayes Model
# =========================================

nb = GaussianNB()

nb.fit(X_train, y_train)
```

```python
# =========================================
# 18. Naive Bayes Prediction
# =========================================

nb_pred = nb.predict(X_test)
```

```python
# =========================================
# 19. Naive Bayes Evaluation
# =========================================

print("Accuracy:",
      accuracy_score(y_test, nb_pred))

print("Precision:",
      precision_score(y_test, nb_pred))

print("Recall:",
      recall_score(y_test, nb_pred))

print("F1 Score:",
      f1_score(y_test, nb_pred))
```

```python
# =========================================
# 20. Model Comparison Table
# =========================================

results = pd.DataFrame({

'Model':[
'Logistic Regression',
'Decision Tree',
'KNN',
'Naive Bayes'
],

'Accuracy':[

accuracy_score(y_test,lr_pred),

accuracy_score(y_test,dt_pred),

accuracy_score(y_test,knn_pred),

accuracy_score(y_test,nb_pred)

],

'Precision':[

precision_score(y_test,lr_pred),

precision_score(y_test,dt_pred),

precision_score(y_test,knn_pred),

precision_score(y_test,nb_pred)

],

'Recall':[

recall_score(y_test,lr_pred),

recall_score(y_test,dt_pred),

recall_score(y_test,knn_pred),

recall_score(y_test,nb_pred)

],

'F1_Score':[

f1_score(y_test,lr_pred),

f1_score(y_test,dt_pred),

f1_score(y_test,knn_pred),

f1_score(y_test,nb_pred)

]

})

results
```

```python
# =========================================
# 21. Best Model
# =========================================

results.sort_values(
    by='F1_Score',
    ascending=False
)
```

```python
# =========================================
# 22. Final Conclusion
# =========================================

print(
"Baseline model training completed successfully."
)
```
